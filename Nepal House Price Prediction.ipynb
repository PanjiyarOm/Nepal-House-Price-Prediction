```python
import re

def convert_to_aana(value):
    # Conversion constants
    ROPANI_TO_AANA = 16
    KATTHA_TO_AANA = 20
    BIGHA_TO_AANA = 320
    SQ_FEET_TO_AANA = 342.25
    SQ_METER_TO_AANA = 31.79  # Approximate, depends on the specific region
    HAAT_TO_SQ_FEET = 30.25   # Assuming 1 Haat = 1.5 feet, so 1 Haat^2 = 2.25 Sq. Feet
    
    # Handle cases with 'Ropani' or 'Sq. Feet'
    if 'Ropani' in value:
        match = re.search(r'(\d+\.?\d*)', value)
        if match:
            ropani_value = float(match.group(1))
            return ropani_value * ROPANI_TO_AANA  # 1 Ropani = 16 Aanas
    elif 'Kattha' in value:
        match = re.search(r'(\d+\.?\d*)', value)
        if match:
            kattha_value = float(match.group(1))
            return kattha_value * KATTHA_TO_AANA  # 1 Kattha = 20 Aanas
    elif 'Bigha' in value:
        match = re.search(r'(\d+\.?\d*)', value)
        if match:
            bigha_value = float(match.group(1))
            return bigha_value * BIGHA_TO_AANA  # 1 Bigha = 320 Aanas
    elif 'Sq. Feet' in value or 'Sq.Feet' in value:
        match = re.search(r'(\d+\.?\d*)', value)
        if match:
            sq_feet_value = float(match.group(1))
            return sq_feet_value / SQ_FEET_TO_AANA  # Conversion of Sq. Feet to Aanas
    elif 'Sq. Meter' in value:
        match = re.search(r'(\d+\.?\d*)', value)
        if match:
            sq_meter_value = float(match.group(1))
            return sq_meter_value / SQ_METER_TO_AANA  # Conversion of Sq. Meter to Aanas
    elif 'Haat' in value:
        dimensions = list(map(float, re.findall(r'(\d+\.?\d*)', value)))
        if len(dimensions) == 2:
            area_in_sq_feet = dimensions[0] * dimensions[1] * HAAT_TO_SQ_FEET
            return area_in_sq_feet / SQ_FEET_TO_AANA  # Conversion of Haat to Aanas
    elif 'Dhur' in value:
        match = re.search(r'(\d+\.?\d*)', value)
        if match:
            dhur_value = float(match.group(1))
            # 1 Dhur = 68.06 Sq. Feet, thus converting to Aanas
            return dhur_value * 68.06 / SQ_FEET_TO_AANA  
    elif '-' in value:
        # For format like '3-5-0-0 Ropani', '0-9-5-0 Kattha', or '0-21-0-0 Aana'
        parts = list(map(int, re.findall(r'\d+', value)))
        ropani, aana, paisa, dam = (parts + [0] * (4 - len(parts)))[:4]  # Ensure list has exactly 4 elements
        unit_value = 0
        if 'Ropani' in value:
            unit_value = ropani * ROPANI_TO_AANA
        elif 'Kattha' in value:
            unit_value = ropani * KATTHA_TO_AANA
        else:  # Default to Aana if no unit is specified
            unit_value = ropani * ROPANI_TO_AANA
        return unit_value + aana + paisa / 4 + dam / 16
    else:
        match = re.search(r'(\d+\.?\d*)', value)
        if match:
            return float(match.group(1))

    # If no match or known format is found, return a default or error value (e.g., 0 or None)
    return None

data['Area'] = data['Area'].apply(convert_to_aana)

# # Convert each entry and print the result
# converted_data = [convert_to_aana(entry) for entry in data['Area']]

# for i, original in enumerate(data['Area']):
#     print(f"Original: {original} -> Converted: {converted_data[i] if converted_data[i] is not None else 'Error'} Aana")
```

```python
def process_location(location):
    parts = [part.strip() for part in location.split(',')]
    if len(parts) < 3:
        return np.nan  # Not enough parts to process
    if parts[0].lower() == 'na':
        return parts[1]  # Use the second part if the first is 'NA'
    return parts[0]  # Use the first part otherwise

# Apply the processing function to the 'Location' column
data['Address'] = data['Address'].apply(process_location)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.impute import SimpleImputer
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

# Applying linear regression

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown='ignore', sparse_output=False),['Address','City','Face']), remainder='passthrough')
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans, imputer, scaler, lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

# Lasso

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso=pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lasso)
```
